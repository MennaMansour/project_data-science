```python
release_trend = merged_df['release_year'].value_counts().sort_index()

# Create the Plotly figure
fig = px.line(x=release_trend.index, y=release_trend.values, 
              labels={'x': 'Year', 'y': 'Number of Movies'},
              title='Number of Movies Released Over the Years')

# Show the figure
fig.show()
```

```python
# Calculate average ratings and popularity per year
average_ratings_per_year = merged_df.groupby('release_year')['vote_average'].mean().reset_index()
average_popularity_per_year = merged_df.groupby('release_year')['popularity'].mean().reset_index()
import plotly.express as px

# Plot average ratings over the years
fig_ratings = px.line(average_ratings_per_year, x='release_year', y='vote_average',
                      title='Average Ratings Over the Years',
                      labels={'release_year': 'Release Year', 'vote_average': 'Average Rating'})
fig_ratings.show()

# Plot average popularity over the years
fig_popularity = px.line(average_popularity_per_year, x='release_year', y='popularity',
                         title='Average Popularity Over the Years',
                         labels={'release_year': 'Release Year', 'popularity': 'Average Popularity'})
fig_popularity.show()

```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces for average ratings
fig.add_trace(
    go.Scatter(x=average_ratings_per_year['release_year'], y=average_ratings_per_year['vote_average'], 
               name='Average Rating', line=dict(color='blue')),
    secondary_y=False,
)

# Add traces for average popularity
fig.add_trace(
    go.Scatter(x=average_popularity_per_year['release_year'], y=average_popularity_per_year['popularity'], 
               name='Average Popularity', line=dict(color='red')),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text='Trends in Average Ratings and Popularity Over the Years'
)

# Set x-axis title
fig.update_xaxes(title_text='Release Year')

# Set y-axes titles
fig.update_yaxes(title_text='Average Rating', secondary_y=False)
fig.update_yaxes(title_text='Average Popularity', secondary_y=True)

# Show the plot
fig.show()
```

```python
# Group by 'release_year' and aggregate 'vote_average', 'popularity', and 'original_title'
grouped_df =merged_df.groupby('release_year').agg({
    'vote_average': 'mean',
    'popularity': 'mean',
    'original_title': lambda x: ', '.join(x)
}).reset_index()

# Check the structure
grouped_df.head()
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces for average ratings
fig.add_trace(
    go.Scatter(x=grouped_df['release_year'], y=grouped_df['vote_average'], 
               name='Average Rating', line=dict(color='blue'),
               hovertext=grouped_df['original_title'], hoverinfo='text+y'),
    secondary_y=False,
)

# Add traces for average popularity
fig.add_trace(
    go.Scatter(x=grouped_df['release_year'], y=grouped_df['popularity'], 
               name='Average Popularity', line=dict(color='red'),
               hovertext=grouped_df['original_title'], hoverinfo='text+y'),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text='Trends in Average Ratings and Popularity Over the Years with Movie Titles'
)

# Set x-axis title
fig.update_xaxes(title_text='Release Year')

# Set y-axes titles
fig.update_yaxes(title_text='Average Rating', secondary_y=False)
fig.update_yaxes(title_text='Average Popularity', secondary_y=True)

# Show the plot
fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a merged_dfFrame named 'merged_df' containing the relevant columns: 'popularity', 'vote_average', and 'vote_count'

# Create a pairplot to visualize the relationships
sns.pairplot(merged_df[['popularity', 'vote_average', 'vote_count']])
plt.show()

# Calculate the correlation matrix
correlation_matrix = merged_df[['popularity', 'vote_average', 'vote_count']].corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
